### Weird join types: Floats
Number: Without any parameters, will be tried

```python
query = f"""
  SELECT column_type_original, ANY_VALUE(column_base_type), COUNT(*) as cnt, list_distinct(list(repo_id))  as repos
  FROM column_usages
  JOIN columns ON column_usages.column_id = columns.id
  JOIN queries on queries.id = column_usages.query_id
  WHERE
    column_usages.operator_type = 'JOIN'
    AND columns.column_base_type = 'Float'
  GROUP BY column_type_original ORDER BY cnt DESC
  """
con.execute(query).df()
```
